# Data Visualisation

## Demographics

### Gender

```r
participants %>%
group_by(GENDER) %>%
summarise(number = n()) %>%
ggplot(mapping = aes(x = GENDER, y = number, fill = GENDER))+
geom_bar(
    stat="identity"
) +
geom_text(aes(label=number), vjust=1.6, color="black", size=3.5) +
theme(legend.position="none") +
theme_minimal() +
theme(legend.position="none") +
labs(title = "Participants Gender") +
xlab("Gender") +
ylab("Number")
```

#### Age

```r
participants %>%
select(AGE) %>%
skim()

participants %>%
ggplot(mapping = aes(y = AGE)) + 
geom_boxplot()
```

#### WPM By Experience

```r
participants %>%
ggplot(mapping = aes(y = AVG_WPM, fill = EXPERIENCED)) +
geom_boxplot()+
labs(title = "WPM grouped by experience") +
ylab("WPM")
```

#### WPM BY CONDITIONS

```r
tests %>% 
select(CONDITION, AVG_WPM) %>%
ggplot(mapping = aes(y = AVG_WPM, fill = CONDITION)) +
geom_boxplot()+
labs(title = "WPM grouped by condition") +
ylab("WPM")
```

#### WPM by Exprience at each Condition

```r
tests %>% 
select(CONDITION, EXPERIENCED, AVG_WPM) %>%
ggplot(mapping = aes(y = AVG_WPM, fill = EXPERIENCED)) +
facet_wrap(~CONDITION)+
geom_boxplot()+
labs(title = "WPM grouped by experience and condition") +
ylab("WPM")


tests %>%
select(AVG_WPM, CONDITION, EXPERIENCED) %>%
group_by(EXPERIENCED,CONDITION) %>%
summarise_all(c("Mean" = mean, "Median" = median, "SD" = sd, "Min" = min, "Max" = max)) %>%
as_tibble() %>%
mutate(across(-c(CONDITION, EXPERIENCED), round, 3)) %>%
arrange(CONDITION)
```

### Error rate

##### Error rate X Condition X Experience

```r
tests %>% 
select(CONDITION, EXPERIENCED, AVG_T_ER) %>%
ggplot(mapping = aes(y = AVG_T_ER, fill = EXPERIENCED)) +
facet_wrap(~CONDITION)+
geom_boxplot()+
labs(title = "Total error rate grouped by experience and condition") +
ylab("WPM")


tests %>%
select(AVG_T_ER, CONDITION, EXPERIENCED) %>%
group_by(EXPERIENCED,CONDITION) %>%
summarise_all(c("Mean" = mean, "Median" = median, "SD" = sd, "Min" = min, "Max" = max)) %>%
as_tibble() %>%
mutate(across(-c(CONDITION, EXPERIENCED), round, 3)) %>%
arrange(CONDITION)
```

```r
tests %>%
select(CONDITION, EXPERIENCED, contains("_ER"))%>%
pivot_longer(
    cols = contains("_ER"),
    names_to = "ER_Type", 
    values_to = "result"
)%>%
ggplot(
    mapping = aes(x = EXPERIENCED, y = result, fill = CONDITION)
)+
geom_boxplot()+
facet_wrap(~ER_Type)
```

AVG_ER: Not corrected errors rate,
AVG_C_ER: Corrected errors rate,
AVG_T_ER: Corrected and Uncorected Error rate
